Первые два предложения содержат информацию о дате, а также целей признак. При этом в конце пресс-релиза присутствует некая общая информаци об организации. Удалим эти части из текстов. 

```python
def clear_text_sentences(text: str) -> str:
    """
    Функция принимание на вход текст. Возвращает текст без первых двух предложений и без куска текста, следующего после предлоежния-маркера.
    """
    #check_sent = 'Контакты для СМИ: pr@raexpert.ru,  тел.: +7 (495) 225-34-44.'
    sentences = text.split('\n')
    count = 0
    for i_sent in sentences:
        count += 1
        if 'Контакты для СМИ' in i_sent:
            break
    #index = sentences.index(check_sent)
    return sentences[2:count-1]
```

```python
df3['clear_text'] = df3['press_release_text'].progress_apply(clear_text_sentences)
```

```python
def clear_text_sentences_akra(text: str) -> str:
    """
    Функция принимание на вход текст. Возвращает текст без первых двух предложений и без куска текста, следующего после предлоежния-маркера.
    """
    #check_sent = 'Контакты для СМИ: pr@raexpert.ru,  тел.: +7 (495) 225-34-44.'
    sentences = text.split('\n')
    start = 0
    end = 0
    count = 0
    for i_sent in sentences:
        count += 1
        if 'Ключевые факторы оценки' in i_sent:
            start = count
        elif 'КОМПОНЕНТЫ РЕЙТИНГА ПО МЕЖДУНАРОДНОЙ ШКАЛЕ' in i_sent:
            end = count
            break
    #index = sentences.index(check_sent)
    return sentences[start:end-1]
```

```python
def clear_text_sentences_nra(text: str) -> str:
    """
    Функция принимание на вход текст. Возвращает текст без первых двух предложений и без куска текста, следующего после предлоежния-маркера.
    """
    sentences = text.split('\n')
    count = 0
    for i_sent in sentences:
        count += 1
        if 'Уровень рейтинга ограничивается' in i_sent:
            break
    return sentences[count:]
```
